```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Supaya plot langsung tampil (jika menggunakan Jupyter Notebook)
%matplotlib inline
```

```python
# 1. Membaca dataset
df = pd.read_csv(r'C:\Users\OPTION\Downloads\Python DA\Projects 27_Spotify-&-Youtube-Content-Trends-Analysis-2023\Spotify_Youtube.csv')
df
```

```python
# Menampilkan informasi dasar
print("Dimensi awal dataset:", df.shape)
```

```python
df.info()
df.describe()
```

```python
#########################
# 1. Membaca dan Mendeteksi Data
#########################

# Cetak informasi awal dataset
print("Dimensi dataset sebelum cleaning:", df.shape)
print("\nInfo dataset:")
print(df.info())
print("\nStatistik deskriptif:")
print(df.describe())
```

```python
# Deteksi data duplikat
duplicate_count = df.duplicated().sum()
print("\nJumlah baris duplikat:", duplicate_count)

# Deteksi nilai kosong (missing values)
missing_values = df.isnull().sum()
print("\nNilai kosong di tiap kolom:\n", missing_values)
```

```python
#########################
# 2. Data Cleaning: Menghapus duplikat, missing values, dan outliers
#########################

# 2.1 Hapus baris duplikat
df_cleaned = df.drop_duplicates()

# 2.2 Hapus baris dengan missing values
df_cleaned = df_cleaned.dropna()
```

```python
# 2.3 Deteksi dan hapus outliers menggunakan metode IQR untuk kolom numerik
# Kita akan mengiterasi semua kolom numerik dan menghapus baris yang memiliki nilai outlier.
numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    # Tetapkan batas bawah dan atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Filter data agar nilai berada di antara batas bawah dan atas
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

```python
# Perbandingan dimensi data sebelum dan sesudah cleaning
print("\nDimensi dataset setelah cleaning:", df_cleaned.shape)
```

```python
#########################
# 3. Visualisasi Histogram dan Box Plot
#########################

# Histogram untuk kolom 'Views'
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Views'], bins=30, kde=True, color='skyblue')
plt.title("Histogram of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.show()

# Box plot untuk kolom 'Likes'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_cleaned['Likes'], color='lightgreen')
plt.title("Box Plot of Likes")
plt.xlabel("Likes")
plt.show()
```

```python
#########################
# 4. Menyimpan Dataset yang Telah Dibersihkan
#########################

# Simpan dataset bersih ke file baru
cleaned_filename = "Spotify_Youtube-CLEANED.csv"
df_cleaned.to_csv(cleaned_filename, index=False)
print("\nDataset yang sudah dibersihkan telah disimpan sebagai:", cleaned_filename)
```

```python
#########################
# 5. Data Visualization: Metrik dan Chart
#########################

# Hitung metrik-metrik
total_contents = len(df_cleaned)
total_artists = df_cleaned['Artist'].nunique()
total_albums = df_cleaned['Album'].nunique()
total_channels = df_cleaned['Channel'].nunique()
total_likes = df_cleaned['Likes'].sum()
total_comments = df_cleaned['Comments'].sum()

print("\n===== Metrik =====")
print("Total Contents :", total_contents)
print("Total Artists  :", total_artists)
print("Total Albums   :", total_albums)
print("Total Channels :", total_channels)
print("Total Likes    :", total_likes)
print("Total Comments :", total_comments)
```

```python
# Pie Chart: Persentase Album Types
plt.figure(figsize=(8, 8))
album_type_counts = df_cleaned['Album_type'].value_counts()
plt.pie(album_type_counts, labels=album_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribusi Tipe Album")
plt.show()
```

```python
# Pie Chart: Persentase Licensed
plt.figure(figsize=(8, 8))
licensed_counts = df_cleaned['Licensed'].value_counts()
plt.pie(licensed_counts, labels=licensed_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribusi Konten Licensed")
plt.show()
```

```python
# Mengelompokkan data berdasarkan Track dan menghitung total Likes
track_likes = df_cleaned.groupby("Track")["Likes"].sum().reset_index()

# Mengurutkan berdasarkan Likes (ranking)
track_likes = track_likes.sort_values("Likes", ascending=False).head(20)  # Menampilkan 20 track teratas

# Plot Line Chart
plt.figure(figsize=(15, 7))
sns.lineplot(x="Track", y="Likes", data=track_likes, marker='o', linestyle='-', color='b')

# Customisasi
plt.title("Ranking Tracks Berdasarkan Counts Likes", fontsize=14)
plt.xlabel("Tracks", fontsize=12)
plt.ylabel("Total Likes", fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.grid(True)
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Mengelompokkan data berdasarkan Track dan menghitung total Komentar
track_likes = df_cleaned.groupby("Track")["Comments"].sum().reset_index()

# Mengurutkan berdasarkan Komentar (ranking)
track_likes = track_likes.sort_values("Comments", ascending=False).head(20)  # Menampilkan 20 track teratas

# Plot Line Chart
plt.figure(figsize=(15, 7))
sns.lineplot(x="Track", y="Comments", data=track_likes, marker='o', linestyle='-', color='b')

# Customisasi
plt.title("Ranking Tracks Berdasarkan Counts Comments", fontsize=14)
plt.xlabel("Tracks", fontsize=12)
plt.ylabel("Total Comments", fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.grid(True)
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Mengelompokkan data berdasarkan Track dan menghitung total Views
track_likes = df_cleaned.groupby("Track")["Views"].sum().reset_index()

# Mengurutkan berdasarkan Views (ranking)
track_likes = track_likes.sort_values("Views", ascending=False).head(20)  # Menampilkan 20 track teratas

# Plot Line Chart
plt.figure(figsize=(15, 7))
sns.lineplot(x="Track", y="Views", data=track_likes, marker='o', linestyle='-', color='b')

# Customisasi
plt.title("Ranking Tracks Berdasarkan Counts Views", fontsize=14)
plt.xlabel("Tracks", fontsize=12)
plt.ylabel("Total Views", fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.grid(True)
plt.tight_layout()

# Menampilkan plot
plt.show()
```
